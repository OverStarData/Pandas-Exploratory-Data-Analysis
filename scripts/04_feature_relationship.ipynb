# 04- Feature Relationship

```python
# import libs

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read data and set rows & columns options
df = pd.read_csv('/Users/deadman/Documents/Pandas-Exploratory Data Analysis/data/coaster_db.csv')

plt.style.use('ggplot')

pd.set_option("display.max_rows",200)
pd.set_option("display.max_columns",200)
```

```python
df.head(2)
```

```python
df.plot(kind="scatter", x='speed_mph', y='height_ft', title="Speed VS Height")

plt.show()
```

```python

sns.scatterplot(
    data=df,
    x='speed_mph',
    y='height_ft',
)
plt.show()
```

```python
# hue is change color based on other values

sns.scatterplot(
    data=df,
    hue="year_introduced",
    x='speed_mph',
    y='height_ft',
)
plt.show()
```

```python
sns.pairplot(
    data=df,
    hue='Type_Main',
    vars=[
        'year_introduced',
        'speed_mph',
        'height_ft'
    ]     
)
plt.show()
```

```python
df_corr = df[[
        'year_introduced',
        'speed_mph',
        'height_ft'
    ]  ].dropna().corr()
```

```python
df_corr
```

```python
sns.heatmap(df_corr,annot=True)
plt.show()
```
